## Matplotlib for Plotting Data

```python
#Step1- import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

#Step2- Prepare Data
x=['English', 'Maths','Language', 'Social']
y=[20,30,40,50]

#Step3- Setup Plot 
fig, ax= plt.subplots(figsize=(3,3)) #width, height of the plot

#Step4 - Plot data
ax.plot(x,y)

#Step5 - Customize plot
ax.set(title='Sample Plot',
      xlabel= 'Subjects',
      ylabel='Marks')

#Step6- save your work
fig.savefig('sample.png')

plt.show()
```

### Bar Plot (Directly from a dictionary)

```python
student_record= {"Mathematics": 46,
                "English" : 44,
                "Physics": 45,
                "Language": 40}
fig, ax= plt.subplots(squeeze=True,figsize=(4,4))
ax.bar(student_record.keys(),student_record.values(),width=0.3)
ax.set(title= "Student Records",
      ylabel= "Marks(out of 50)",
      xlabel= "Subjects");
```

```python
student_record= {"Mathematics": 46,
                "English" : 44,
                "Physics": 45,
                "Language": 40}
fig, ax= plt.subplots(squeeze=True,figsize=(4,4))
ax.barh(list(student_record.keys()),list(student_record.values()),height=0.5)
ax.set(title= "Student Records",
      ylabel= "Marks(out of 50)",
      xlabel= "Subjects");
```

### Subplots

#### Option1

```python
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,
                                         ncols=2,
                                         figsize=(10,5))
ax1.plot(x,x/3)
ax2.scatter(x,np.sin(x))
ax3.bar(student_record.keys(),student_record.values())
ax4.hist(x/2);
```

#### Option 2

```python
fig, ax= plt.subplots(nrows=2,
                     ncols=2,
                     figsize=(10,5))
ax[0,0].plot(x,x/3)
ax[0,1].scatter(np.random.random(10),np.random.random(10))
ax[1,0].bar(student_record.keys(),student_record.values())
ax[1,0].set_ylim([0,50])
ax[1,1].hist(x)
```

### Which method to use when for plotting?
* When making something quick, use pyplot
* When making something advanced, use matplotlib OO method i.e. fig, axis etc

```python
over_50= heartDf[heartDf['age']>50]
over_50
```

```python
#Pyplot method
over_50.plot.scatter(x='age',y='chol',c='target');
```

```python
#Matlab OO Method with PyPlot method
fig, ax= plt.subplots()
over_50.plot.scatter('age','chol',c='target',ax=ax)
ax.set(title= "Patients over 50",
      xlabel="Age",
      ylabel="Cholestrol level");
```

```python
### OO method from scratch
fig, ax= plt.subplots()
scatter= ax.scatter(x=over_50['age'],
                   y=over_50['chol'],
                   c=over_50['target'])
ax.set(xlabel='age',
      ylabel='cholestrol level',
      title='Cholestrol Levels of people above age 50');

## Adding a legend
ax.legend(*scatter.legend_elements(),title="Target Legend");

#Adding a mean line
ax.axhline(over_50['chol'].mean(),
          linestyle= '--')
```

```python
## Creating Subplots
fig, ax = plt.subplots(nrows=2,figsize=(10,10),sharex=True)
scatter1=ax[0].scatter(x=over_50['age'],y=over_50['chol'],c=over_50['target'])
ax[0].set(title='Cholestrol VS Age',
                 ylabel='Cholestrol')
ax[0].legend(*scatter1.legend_elements(),title='Target')
ax[0].axhline(over_50['chol'].mean(),linestyle='--')

scatter2=ax[1].scatter(x=over_50['age'],y=over_50['thalach'],c=over_50['target'])
ax[1].set(title='Thalach VS Age',
         xlabel='Age',
         ylabel='Thalach')
ax[1].legend(*scatter2.legend_elements(),title='Target')
ax[1].axhline(over_50['thalach'].mean(),linestyle='--');

#Adding a title to whole figure
fig.suptitle("Heart Disease Analysis",fontsize=16, fontweight='bold');
```
